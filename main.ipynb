```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data.csv",
                 sep=",",
                 decimal=".")
df.set_index("Year",
             inplace=True)
df.drop("Total",
        axis=1,
        inplace=True)
df
```

```python
# Plot the area graph
ax = df.plot(kind='area', stacked=True, figsize=(10, 6))

# Set labels and title
# ax.set_ylabel('Energy Production')
ax.set_xlabel('Year')
ax.set_title('Energy Production (EJ)')

target_year = 1840
text = "Industry 1.0"
target_value = df.loc[target_year, :].sum()
ax.annotate(text, xy=(target_year, target_value),
            xytext=(target_year - 10, target_value + 10)
            )

target_year = 1900
text = "Industry 2.0"
target_value = df.loc[target_year, :].sum()
ax.annotate(text, xy=(target_year, target_value),
            xytext=(target_year - 10, target_value + 10)
            )

target_year = 1970
text = "Industry 3.0"
target_value = df.loc[target_year, :].sum()
ax.annotate(text, xy=(target_year, target_value),
            xytext=(target_year - 10, target_value + 10),
            )


# Show the plot
plt.show()
```
