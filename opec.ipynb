```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def generate_graph(data:dict, title:str, x_axis:str, y_axis:str) -> None:
    df = pd.DataFrame(data)
    # Plot the area graph
    # https://matplotlib.org/stable/users/explain/colors/colormaps.html
    descriptor = [key for key in data.keys() if type(key)==str][0]
    ax = df.set_index(descriptor).T.plot(kind='area', stacked=True, colormap='tab20b', figsize=(12, 6))
    plt.title(title)
    plt.xlabel(x_axis)
    plt.ylabel(y_axis)
    plt.legend(title=descriptor)
    plt.show()
```
